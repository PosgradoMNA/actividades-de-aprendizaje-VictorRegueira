### Imputación de datos

La imputación de datos requiere de usar distintos tipos de estrategias. Para este caso en nuestro DataFrame tenemos variables categóricas y numéricas, por lo tanto, cada una la vamos a tratar con estrategias distintas  para no afectar nuestros datos

```python
lista_cat = list(df_imp.iloc[:,1:11])
lista_cat.remove('age')
```

```python
lista_cat
```

DataFrame categóricas

```python
df_imp_cat = df_imp[lista_cat]
df_imp_cat.head(2)
```

DataFrame numéricas

```python
df_imp_num = df_imp.drop(df_imp_cat.columns,axis=1)
df_imp_num.drop(axis=1,columns='Y',inplace=True)
df_imp_num.head(2)
```

```python
lista_cat = df_imp_cat.columns
lista_num = df_imp_num.columns
lista_total = [*lista_cat, *lista_num,'Y']
lista_total
```

Empecemos haciendo la imputación de los datos usando el SimpleImputer de la librería de ScickitLearn. De igual forma, usaremos el Pipeline para definir la entrada de nuestro datos y ajustarlos

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

#class sklearn.impute.SimpleImputer(*, missing_values=nan, strategy='mean', fill_value=None, verbose='deprecated', copy=True, add_indicator=False)

cat_pipeline = Pipeline(steps = [
    ("imputer", SimpleImputer(strategy='most_frequent'))
])

num_pipeline = Pipeline(steps = [
    ("imputer", SimpleImputer(strategy='median'))
])
```

Usaremos el ColumnTransformer para crear un DataFrame con todos los datos completos, es decir, variables_categóricas + variables_numéricas

```python
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(transformers=[
    ("cat", cat_pipeline, lista_cat),
    ("num", num_pipeline, lista_num)
], remainder='passthrough');
```

```python
df_imp = pd.DataFrame(transformer.fit_transform(df_imp),columns=lista_total)
```

Al haber agregado la variable de salida Y, por fines prácti

```python
df_imp.dropna(inplace=True)
```

Realizamos una copia de los DataFrames para mantener sus valores antes de realizar cualquier tipo de Normalización, así podremos comparar los datos

## 7 Reduce las dimensiones con PCA, si consideras necesario. 

* PCA puede ayudarnos a mejorar el rendimiento a un costo muy bajo de precisión del modelo de aprendizaje de maquina, dado que el rendimiento del algoritmo depende de la dimensión de los datos, y si el modelo se ejecuta en datos de dimensiones muy altas, puede funcionar muy lentamente, o incluso fallar, y requerir de recursos de servidores significativos.

* PCA también nos permite visualizar datos y permitir la inspección de algoritmos de agrupamiento/clasificación.

* Finalmente, otros beneficios de PCA incluyen la reducción del ruido en los datos, la selección de características (hasta cierto punto) y la capacidad de producir características independientes y no correlacionadas de los datos.  

```python
from sklearn.decomposition import PCA
pcs_imp_a = PCA() #Creamos variable pcs que almacena el método PCA().
pcs_sna_b = PCA()

pcs_imp = pcs_imp_a.fit_transform(scaled_df_imp) ##variable pcs_t almacena los valores aplicados de la transformación de pcs
pcs_sna = pcs_sna_b.fit_transform(scaled_df_sna)
```

Crearemos una variable con las etiquetas de cada Componente Principal (PC)

```python
pcs_labels = [f'PC{i + 1}' for i in range(len(scaled_df_imp.columns))] 
pcs_labels
```

Crearemos una variable que almacenará el DataFrame con nuestros componentes en relación de la Desviación estándar, Proporción de varianza y Proporción Acumulada

```python
pcsSummary_df_imp = pd.DataFrame({'Desviación estándar': np.sqrt(pcs_imp_a.explained_variance_),
                             'Proporción de varianza': pcs_imp_a.explained_variance_ratio_,
                             'Proporción Acumulada': np.cumsum(pcs_imp_a.explained_variance_ratio_)
                             })

pcsSummary_df_imp = pcsSummary_df_imp.T
pcsSummary_df_imp.columns = pcs_labels
pcsSummary_df_imp.round(4)
```

Lo mismo para el DataFrame donde eliminamos los valores nulos

```python
pcsSummary_df_sna = pd.DataFrame({'Desviación estándar': np.sqrt(pcs_sna_b.explained_variance_),
                             'Proporción de varianza': pcs_sna_b.explained_variance_ratio_,
                             'Proporción Acumulada': np.cumsum(pcs_sna_b.explained_variance_ratio_)
                             })

pcsSummary_df_sna = pcsSummary_df_sna.T
pcsSummary_df_sna.columns = pcs_labels
pcsSummary_df_sna.round(4)
```

Aquí observamos como al haber aplicado el StandardScaler, todas nuestras variables están centralizadas de acuerdo al valor de su moda en cero, por eso todos mantienen el porcentaje de varianza similar

```python
total_var = scaled_df_imp.var().sum()
pd.DataFrame({
"Porcentaje Varianza": (scaled_df_imp.var()/ total_var) * 100,
"Porcentaje Varianza Acumulado": (scaled_df_imp.var().cumsum() / total_var) *100
})
```

Si analizamos este % de varianza en los componentes principales,ahora vemos cómo tiene un mayor impacto los componentes yendo del primer componente con más % de varianza, hasta llegar a 100%

```python
pcsSummary_df_imp = pd.DataFrame({
'% varianza explicada': np.round(pcs_imp_a.explained_variance_ratio_,4) * 100,
'% varianza acumulada': np.cumsum(pcs_imp_a.explained_variance_ratio_) * 100
},
index = pcs_labels)
pcsSummary_df_imp
```

Lo mismo sucede con el DataFrame de los valores nulos eliminados

```python
pcsSummary_df_sna = pd.DataFrame({
'% varianza explicada': np.round(pcs_sna_b.explained_variance_ratio_,4) * 100,
'% varianza acumulada': np.cumsum(pcs_sna_b.explained_variance_ratio_) * 100
},
index = pcs_labels)
pcsSummary_df_sna
```

Hasta este punto podemos ver que el haber imputado los valores nulos y haber elimando los valores nulos nos proporcionan valores muy similares.

> También, concluímos que con **5 componentes** ya cumplimos con el **74.9%** de varianza acumulada, cumpliendo con los términos solicitados por la tarea

```python
pcsComponents_df_imp = pd.DataFrame(pcs_imp_a.components_.round(4),#datos pcs.components nos da los pesos de cada uno de los componentes
                                columns=pcs_labels,#columnas Son 14 columnas, las cuales hacen referencia a cada uno de los componentes: PC1, PC2...PC10
                                index=scaled_df_imp.columns #índice (ID) Las filas son 10 también, las cuales hacen referencia a las variables originales
                                )
pcsComponents_df_imp
```

```python
pcsComponents_df_imp.abs().idxmax() #Nos interesa buscar los máximos por filas, no por columnas. Por eso usamos idxmax
#Si quisieramos especidicar en una sola columna, llamaríamos el nombre del componente (PCA1)
```

> De igual forma, aquí vemos la importancia de las variables por cada uno de los componentes. Vemos que las variables paid tienen mayor impacto en la varianza para cada componente. Así que, deberíamos considerar estos primeros 5 componentes para escoger variables que tengan más relevancia en la evaluación y análisis de las variables

```python
pcsComponents_df_imp.PC5.abs().nlargest(5)
```

La variable del PC5 se repite con el del PC4, por lo que podemos escoger la variable que esté en segundo lugar: **paid_jun**
